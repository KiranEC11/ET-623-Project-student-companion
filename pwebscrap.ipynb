```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
def skillScrapperClass(html_tag, course_case, p_class, tag_class, div_class=None):
    for i in range(1,84):
        url = "https://www.coursera.org/courses?page=" +str(i)
        # print(url)
        #url = "https://www.coursera.org/search?query=data%20science&page=" +str(i)

        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        elements = soup.find_all('div', class_=div_class)

        for name in elements:
            skills_tag = name.find('div', class_= tag_class)
            if skills_tag:
                skill = skills_tag.find(html_tag, class_= p_class)
                # print(x)
                if skill:
                    x = skill.get_text()
                    if x:
                        course_case.append(x)
                    else:
                        course_case.append(None)
                else:
                    course_case.append(None)
            else:
                course_case.append(None)
```

```python
def linkScrapperClass(html_tag, course_case, tag_class, div_class=None, sec_class=None):
    for i in range(1,84):
        url = "https://www.coursera.org/courses?page=" +str(i)
        # print(url)
        #url = "https://www.coursera.org/search?query=data%20science&page=" +str(i)

        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        elements = soup.find_all('div', class_=sec_class)


        # print(elements[0])
        for name in elements:
          tag = name.find(html_tag)['href']
          if tag:
              course_case.append('https://www.coursera.org'+tag)
          # else:
          #     course_case.append(None)
```

```python
def autoScrapperClass(html_tag, course_case, tag_class, div_class=None, sec_class=None):
    for i in range(1,84):
        url = "https://www.coursera.org/courses?page=" +str(i)
        # print(url)
        #url = "https://www.coursera.org/search?query=data%20science&page=" +str(i)

        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        elements = soup.find_all('div', class_=div_class)


        # print(elements[0])
        for name in elements:
          tag = name.find(html_tag, class_= tag_class)
          if tag:
              x = tag.get_text()
              # print(x)
              if x:
                  course_case.append(x)
              else:
                  course_case.append(None)
          else:
              course_case.append(None)
```

```python
def insScrapperClass(html_tag, course_case, p_class, tag_class, url, div_class=None):
    url = url
    # print(url)
    #url = "https://www.coursera.org/search?query=data%20science&page=" +str(i)

    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    elements = soup.find('div', class_=div_class)
    for name in elements:
        skills_tag = name.find('div', class_= tag_class)
        if skills_tag:
            skill = skills_tag.find('img')[html_tag]
            if skill:
              course_case.append(skill)
            else:
                course_case.append(None)
        else:
            course_case.append(None)
```

```python
def freeScrapperClass(html_tag, course_case, p_class, tag_class, url, div_class=None):
    url = url
    # print(url)
    #url = "https://www.coursera.org/search?query=data%20science&page=" +str(i)

    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    elements = soup.find('div', class_=div_class)
    for name in elements:
        skills_tag = name.find('span', class_= tag_class)
        if skills_tag:
            skill = skills_tag.find('span', {'data-test': 'enroll-button-label'})
            if skill:
              enroll_text = skill.text
              print(enroll_text)
              course_case.append(enroll_text)
            else:
                course_case.append(None)
        else:
            course_case.append(None)
```

```python
def subScrapperClass(html_tag, course_case, p_class, tag_class, url, div_class=None):
    url = url
    # print(url)
    #url = "https://www.coursera.org/search?query=data%20science&page=" +str(i)

    page = requests.get(url)
    flag=0
    soup = BeautifulSoup(page.content, 'html.parser')
    elements = soup.find_all('div', class_=div_class)
    if len(elements):
      elements = elements[-1]
      for name in elements:
          if name!=None:
            text = name.get_text()
            numerical_value = re.search(r'\d{1,3}(,\d{3})*', text)

            if numerical_value:
                numerical_value = numerical_value.group().replace(',', '')
                numerical_value = int(numerical_value)
                print(numerical_value)
                flag=1
                course_case.append(numerical_value)
            else:
                flag=1
                course_case.append(0)
                print(0)
          else:
              course_case.append(0)
              flag=1
              print(0)
    if flag==0:
        course_case.append(0)
        print(0)
```

```python
def rewScrapperClass(html_tag, course_case, p_class, tag_class, url, div_class=None):
    url = url
    # print(url)
    #url = "https://www.coursera.org/search?query=data%20science&page=" +str(i)

    page = requests.get(url)
    flag=0
    soup = BeautifulSoup(page.content, 'html.parser')
    elements = soup.find('div', class_=div_class)
    if elements:
      skill = elements.find(html_tag, class_= tag_class )
      if skill:
        course_case.append(skill.text)
      else:
        course_case.append('0')
    else:
      course_case.append('0')
```

```python
def desScrapperClass(html_tag, course_case, p_class, tag_class, url, div_class=None):
    url = url
    # print(url)
    #url = "https://www.coursera.org/search?query=data%20science&page=" +str(i)

    page = requests.get(url)
    flag=0
    soup = BeautifulSoup(page.content, 'html.parser')
    elements = soup.find('section', class_= div_class)
    print(elements)
    if elements:
      skill = elements.find('div', class_= tag_class)
      if skill:
        skill_ = skill.find_all('span')
        desc =''
        for i in skill_:
          desc+=i.text
          desc+=','
        course_case.append(desc)
      else:
        course_case.append('')
    else:
      course_case.append('')
```
