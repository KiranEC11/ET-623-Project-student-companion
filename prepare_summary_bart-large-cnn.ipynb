```python
import numpy as np
import pandas as pd
```

```python

# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")
```

```python

# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("summarization", model="facebook/bart-large-cnn")

df = pd.read_csv("notebooks/chatbot/coursera_edx_main_features.csv")
```

```python
df.head()
```

```python
## we need to create a summary of description
from tqdm import tqdm
tqdm.pandas()

def SUMMARY(x):
    if len(x)< 1000:
        return x
    else:
        return pipe(x)


```

```python
## apply fn
short_summary = []
for i,item in tqdm(enumerate(df['Decription']), total=df.shape[0]):
    try:
        short_summary.append(SUMMARY(item))
    except:
        short_summary.append(item)
```

```python
short_summary[1]
```

```python
short_summary_text = []
for item in short_summary:
    if isinstance(item, list) and len(item) > 0 and isinstance(item[0], dict):
        # If item is a list and contains at least one dictionary, extract 'summary_text' from the first dictionary
        summary_text = item[0].get('summary_text', '')
        short_summary_text.append(summary_text)
    elif isinstance(item, str):
        # If item is a string, append it directly to the list
        short_summary_text.append(item)

# Print the result (for verification)
print(len(short_summary_text))
```

```python
df['short summary'] = short_summary_text
df.to_csv("notebooks/chatbot/comparison_ouput_short_desc_bart_large.csv", index=False)
df.head()
```
